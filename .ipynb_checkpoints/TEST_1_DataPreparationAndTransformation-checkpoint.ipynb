## >>> 5.2 - Accurate Timestamp
### !!!!!!!! NOT NECESSARY FOR THIS DATASET

```r
race_2021_clean_df <- data.frame(race_2021_clean_df)

head(as.data.frame(race_2021_clean_df$General.Timestamp), 15)
```

```r
race_2021_clean_df$Calc.Timestamp <- NULL
race_2021_clean_df <- race_2021_clean_df[order(race_2021_clean_df$General.Timestamp),]

initialTimestamp <- race_2021_clean_df[1, 'General.Timestamp']

rowIndex <- 1
for(i in 2:nrow(race_2021_clean_df)){
    tempTimestamp <- race_2021_clean_df[i, 'General.Timestamp']
    time_difference <- difftime(tempTimestamp, initialTimestamp, units='mins')
    
    # if we find time difference of seconds
    if(time_difference == 1)
    {
        data_rows = i - rowIndex
        time_delta = 60 / data_rows
        
        if(time_delta > 1){
            time_delta = 1
        }
        
        for(j in rowIndex: i)
        {
            a <- race_2021_clean_df[rowIndex, 'General.Timestamp'] + (j - rowIndex)  * time_delta
            sprintf("%f", a)
            
            race_2021_clean_df[j, 'Calc.Timestamp'] <- race_2021_clean_df[rowIndex, 'General.Timestamp'] + (j - rowIndex)  * time_delta
        }
            break
        
        rowIndex <- i
        initialTimestamp <- tempTimestamp
    }
    
    if(time_difference > 1){
        data_rows = i-1 - rowIndex
        time_delta = 60 / data_rows
        
        if(time_delta > 1){
            time_delta = 1
        }
        
        for(j in rowIndex: i-1)
        {
            race_2021_clean_df[j, 'Calc.Timestamp'] <- race_2021_clean_df[rowIndex, 'General.Timestamp'] + (j - rowIndex) * time_delta
        }
        rowIndex <- i
        initialTimestamp <- tempTimestamp
    }
}
asdasd <- race_2021_clean_df
```

## >>> 5.3 - Race Cumultative Mileage and Speed
We want to know exact mileage of the car during the race. As the team changed the tyre type just before the race - the speed value showed by gauges was not perfect. Knowing RPM and tyre model we will calculate new linear speed

### Tyre cirumference = 2150 mm
### V_lin [km/h] = RPM[1/s] * Circumference[mm] * 60 / 1 000 000

```r
tyre_circumference = 2150 #mm

# Calculate the instantaneous speed [km/h]
race_2021_clean_df$Calc.Speed <- race_2021_clean_df$General.Rpm * tyre_circumference * 60 / 1000000

# Compare speed measured by the car and our calculation
summary(race_2021_clean_df$Calc.Speed)
summary(race_2021_clean_df$General.Speed)
```

```r
# Get Time Difference between rows of data[s]
race_2021_clean_df$Calc.TimeDiff <- as.numeric(
    (race_2021_clean_df$Calc.Timestamp - shift(n=3L, race_2021_clean_df$Calc.Timestamp)) /3)

# race_2021_clean_df$Calc.TimeDiff <- as.numeric(
#     race_2021_clean_df$General.Timestamp - shift(race_2021_clean_df$General.Timestamp))

race_2021_clean_df$Calc.TimeDiff <- ifelse(race_2021_clean_df$Calc.TimeDiff <= 0.5, race_2021_clean_df$Calc.TimeDiff, 0.5)
race_2021_clean_df$Calc.TimeDiff <- ifelse(race_2021_clean_df$Calc.TimeDiff > 0, race_2021_clean_df$Calc.TimeDiff, 0)


# Get Mileage [s * km / h] covered during TimeDiff
race_2021_clean_df$Calc.Mileage <- abs(
    race_2021_clean_df$Calc.TimeDiff * race_2021_clean_df$Calc.Speed) / 3600

# Get rid of outliers (created when car was swiched off)
race_2021_clean_df$Calc.Mileage <- ifelse(race_2021_clean_df$Calc.Mileage <= 0.3, race_2021_clean_df$Calc.Mileage, 0)
race_2021_clean_df$Calc.Mileage[is.na(race_2021_clean_df$Calc.Mileage)] <- 0

# Get total race Mileage [km]
sum(race_2021_clean_df$Calc.Mileage)
```
